```python
# install if needed
# !pip install pandas_datareader

import pandas as pd
from pandas_datareader import data as pdr

# Pull daily Brent spot prices from FRED
# DCOILBRENTEU = Brent crude spot price, USD per barrel
brent = pdr.DataReader("DCOILBRENTEU", "fred", start="1993-01-01")
brent = brent.rename(columns={"DCOILBRENTEU": "Brent_Price_USD"})

# Resample daily to weekly (week ending Monday to match EIA gas price release day)
# Use mean so the weekly price reflects the full week, not just one day
brent_weekly = brent.resample("W-MON").mean().dropna().reset_index()
brent_weekly = brent_weekly.rename(columns={"DATE": "Date"})

# Save to CSV
brent_weekly.to_csv("brent_crude_weekly_1993_present.csv", index=False)

print(f"Rows: {len(brent_weekly)}")
print(brent_weekly.head())
print(brent_weekly.tail())
```
