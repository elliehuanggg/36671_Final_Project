```python
import pandas as pd
from pandas_datareader import data as pdr

# Pull daily Brent spot prices from FRED
brent = pdr.DataReader("DCOILBRENTEU", "fred", start="1993-01-01")

# Rename column
brent = brent.rename(columns={"DCOILBRENTEU": "Brent_Price_USD"})

# Convert daily to monthly using last available value in each month
brent_monthly = brent.resample("ME").last().dropna().reset_index()

# Save to CSV
brent_monthly.to_csv("brent_crude_monthly_1993_present.csv", index=False)

print(brent_monthly.head())
print(brent_monthly.tail())
```
